3) Use forward finite difference to approximate $f'(0.5)$ with $h=0.001$

````python
def f(x):
    return x**3
h = 0.001
print (f(0.5+h)-f(0.5))/h
````

````python
#3.5 plot f'(x) for h = 0.1 and h =0.001 and analytic one
def analytic(x):
    return 3*x**2
def numericalx(x,h):
    return (f(x+h) - f(x))/h
#magic
def center(x,h):
    return (f(x+h)-f(x-h))/(2*h)

xs = np.linspace(0,1,200)
analytic_ys = [analytic(x) for x in xs]
highh_ys = [numericalx(x,0.1) for x in xs]
smallh_ys = [numericalx(x,0.001) for x in xs]
center_ys = [center(x,0.1) for x in xs]


plt.plot(xs,analytic_ys, label = "analytic")
plt.plot(xs,highh_ys, label = "High h")
plt.plot(xs,smallh_ys, label = "low h",lw = 5,alpha = 0.3)
plt.plot(xs,center_ys, label = "center",lw = 2,alpha = 0.9)
plt.legend()

````

# Newton's method with numerical derivative

1) Write a function
```def newton(f, start, n):```
that try to find solution of $f(x)=0$ given starting point at $x = start$ for n iteration. But this time you don't need to do $f'(x)$ by hands.

Try it on our usual suspect: solving $x = \cos(x)$



````python
def f(x):
    return math.cos(x) - x

def center(x,h):
    return (f(x+h)-f(x-h))/(2*h)

def newton(f,start,n):
    print ("original guess", start)
    for i in range(n):
        start = start - f(start)/center(start,0.1)
        print (i, "updated guess",start)
    return start

newton(f,1,5)
#5 iterations are usually enough
````

5) Write a function that given location $y,x$ and then calculate how "likely" it is that pixel x,y is an edge pixel.

````python
def diff(orig,neigh):
    return np.sum( orig-neigh )**2
def edgy(pic, y, x):
    center = pic[ y, x] #itself
    top = pic[ y - 1, x ]
    bottom = pic[ y + 1, x ]
    left = pic[ y , x -1 ]
    right = pic[ y, x + 1 ]
    
    score = diff(center,top)
    score += diff(center,bottom)
    score += diff(center,left)
    score += diff(center,right)
    return score
````

6) Make an empty 3d array of (ydimension, xdimension, ncolor). Then start putting in each pixel with the size of gradient.

````python

def findEdges(image):
    #print image.shape
    tmp = np.zeros(image.shape)
    ny = tmp.shape[0]
    nx = tmp.shape[1]
    nz = tmp.shape[2]
    for iy in range(1,ny-1):
        for ix in range(1,nx-1):
            score = edgy(image,iy,ix)
            tmp[iy,ix] = np.array([score]*nz)
    return tmp
marioEdges = findEdges(mario)
bugEdges = findEdges(stinkbug)





        
````
