2) Find the polynomial of degree 3 that pass through these four points. Plot to verify.

```python
def f(x):
    return 3*w1(x,2,3,4,5) + w2(x,2,3,4,5) + 2*w3(x,2,3,4,5) + 2*w4(x,2,3,4,5)
def w1(x,x1,x2,x3,x4):
    return ((x-x2)*(x-x3)*(x-x4))/((x1-x2)*(x1-x3)*(x1-x4))

def w2(x,x1,x2,x3,x4):
    return ((x-x1)*(x-x3)*(x-x4))/((x2-x1)*(x2-x3)*(x2-x4))

def w3(x,x1,x2,x3,x4):
    return ((x-x1)*(x-x2)*(x-x4))/((x3-x1)*(x3-x2)*(x3-x4))

def w4(x,x1,x2,x3,x4):
    return ((x-x1)*(x-x2)*(x-x3))/((x4-x1)*(x4-x2)*(x4-x3))
```

```python
xs = np.linspace(2,5,5000)
ys = [f(x) for x in xs]
plt.plot(2,3,"o", color ="b")
plt.plot(3,1,"o", color ="b")
plt.plot(4,2,"o",color ="b")
plt.plot(5,2,"o",color ="b")
plt.plot(xs,ys)
```

1)Use Trapezoid Rule to calculate

$$ \int \limits_{x=1}^{x=3} x^3 \; dx $$

Using 1 piece.

```python
#how to check correctness
def trapz(f,a,b,n):
    edges = np.linspace(a,b,n+1)
    n = float(n)
    h = (b-a)/n
    #sum it up
    return (sum(f(e) for e in edges[1:-1]) + 0.5*(f(a) + f(b)))*h
      
trapz(lambda x: x**3,1,3,1)
```
