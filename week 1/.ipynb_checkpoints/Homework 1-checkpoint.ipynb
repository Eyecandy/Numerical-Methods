1.1) If we estimate $\cos(0.1)$ using 4 **non-zero** terms, what is the bound on the **truncation error**.

```python
def truncError(x,n):
    if (n==0):
        return 1
    n = 2 * (n - 1) + 1 #we get 1 term for free cuz the 7th term is 0
    return x**n/math.factorial(n)
print (truncError(0.1,4))
```

1.2) How many **non-zero** terms do we need to calculate $\cos(0.1)$ such that the bound on the **truncation error** is less than $10^{-7}$

```python
def findTruncErrorLessThan(e):
    term_i = 0
    while (truncError(0.1, term_i) >= e):
        term_i+=1
    return term_i


print (findTruncErrorLessThan(1e-7)) 
```

```python
def truncErrorCalc(x,n):
    n = n + 1
    return ((x)**n)/n
print (truncErrorCalc(1.5,3))
```

3.3) Calculate $\pi$ using taylor approximation for 100 **non-zero** terms. (Note I changed this to 100 terms so you must write some sort of for loop to do it).

```python
def arcTan(x,n):
    alternate = 1
    summ_terms = 0
    for i in range(1,n):
        summ_terms += x**(2*i-1)/float(2*i -1) *alternate 
        alternate = -alternate
    return summ_terms
arcTan(1.0,100) * 4 #arcTan gives pi/4
        
```
