10) Give me the solution to $x=\cos(x)$ that is at most $10^{-6}$ away from the real answer.

```python

def bisection(fromm,to):
    guess = (fromm+to)/float(2)
    if (abs(f(guess)) <= 1e-6):
        return guess
    else:
        if (f(guess) > 0): 
             return bisection(guess,to)
        else:
             return bisection(fromm, guess)
            
ans = bisection(0.2,1.2)     
print "BISECTION solution x = cos(x): " , ans
```

17) Now find out the general formula for next guess given that the current guess is (x_n).

```python
def getM(x):
    return 3*x**2 - 1

def getY(x):
    return x**3-x -1

def getC(x,m,y):
    c = y - m*x
    return c

def getXintercept(x,m):
    y = getY(x)
    c = getC(x,m,y)
    xIntercept = -c/float(m)
    return xIntercept
    
x = 2
m = getM(x)
xIntercept = getXintercept(x,m) #next guess
```
